```python
from tqdm import tqdm
import cv2
import numpy as np
import os

initDirec = r'C:\Users\rb0063\OneDrive - University of Surrey\Desktop\zoon data\Finished Data\All_Volunteers_Crowdsourced\images\test'
targetDirec = r'C:\Users\rb0063\OneDrive - University of Surrey\Desktop\zoon data\Finished Data\All_Volunteers_Crowdsourced\images\train'


repeatcount = 0
toRemove = []
#print(os.listdir(initDirec))
for filename in os.listdir(initDirec):
    if filename.endswith(".jpg") or filename.endswith(".JPG") or filename.endswith(".png") or filename.endswith(".bmp"):
        original = cv2.imread(str(os.path.join(initDirec, filename)))
        if original is None:
            continue
        else:
            for filename2 in tqdm(os.listdir(targetDirec)):
                duplicate = cv2.imread(str(os.path.join(targetDirec, filename2)))
                if filename.endswith(".jpg") or filename.endswith(".JPG") or filename.endswith(".png") or filename.endswith(".bmp"):
                    if filename == filename2:
                        continue
                    elif duplicate is None:
                        continue 
                    else:
                        if original.shape == duplicate.shape:
                            #print("The images have same size and channels")
                            difference = cv2.subtract(original, duplicate)
                            b, g, r = cv2.split(difference)
                            if cv2.countNonZero(b) == 0 and cv2.countNonZero(g) == 0 and cv2.countNonZero(r) == 0:
                                #print("The images are completely Equal")
                                repeatcount += 1
                                #print(repeatcount)
                                toRemove.append(str(filename2))
                    #print(filename, end='\r')
                    #print(filename2, end='\r')


                else:
                    continue
    else:
        continue
print(repeatcount, end='\r')
print("number of repeated images: ", repeatcount)
print(toRemove)
#print(os.path.join(directory, filename))
```
