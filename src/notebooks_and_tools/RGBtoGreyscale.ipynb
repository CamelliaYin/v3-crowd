### RGB to Greyscale Converter
This script converts all images in a directorry into greyscale


```python
import cv2
import os
from PIL import Image
```

```python
directory = r'C:\Users\rb01243\OneDrive - University of Surrey\Documents\GitHub\C-YOLOv3\v3-crowd\data\datasets\Zoon_50min_3perImg_Crowdsourced\images\train'

count = 0 
for file in os.listdir(directory):
    img = cv2.imread(directory+'\\'+file)
    #print(img.shape[2])
    if(len(img.shape)==3):
        imgs = Image.open(directory+'\\'+file).convert('L')
        imgs.save(directory + '\\' + file.split('.')[0] + '.jpg')
        count += 1 
print ('DONE')
print(count, ': Images Converted to Greyscale')

count2 = 0
for file in os.listdir(directory):
    img = cv2.imread(directory+'\\'+file)
    if(len(img.shape)==3):
        count2 += 1
print(count2, ': Images that havent converted correctly')
```
