#### This code removes all labels except for bone loss in a training folder

```python
import os
import pandas as pd
import numpy as np
```

```python

# This code traverses all the training lables in a desired folder and removes all labels except for bone loss (int 5)
# In:
#      curFolder (string) : name of the current folder that needs to be traversed
#      initDirect (string) : directory of the labels folder in the yoloV5 directory
#      ct (int) : current count for either train test or val folders
# Out: 
#      True if successfull
def Traverse (curFolder, initDirect, ct):
    direc = os.getcwd()
    for filename in os.listdir(direc):
        df = pd.read_table(filename, delim_whitespace=True, names=('label', 'X', 'Y', 'height','width'))
        df = df.loc[df['label'] == 5]
        if ct == 0:
            folder = 'train5\\'
        elif(ct==1):
            folder = 'test5\\'
        else:
            folder = 'val5\\'
        finalDir = initDirect+folder+filename
        np.savetxt(finalDir, df.values, fmt='%i %1.6f %1.6f %1.6f %1.6f')
        print(df)
    return(True)

    
```

```python
# directory of the labels you want to change
direct = r'C:\Users\\Ryan\\OneDrive - University of Surrey\\TERM 2\\DISSERTATION\\1.1 Models\\yolov5\\data\\labels\\'
# the folder names of the train test and val folder you want to change
fileNme = ['train2\\', 'test2\\', 'val2\\']
os.chdir(direct)
saveDir = os.getcwd()
# changes the directory for each folder that contains label data
count = 0
for i in fileNme:
    os.chdir(direct + i)
    Traverse(i, direct, count)
    os.chdir(saveDir)
    count += 1
```
